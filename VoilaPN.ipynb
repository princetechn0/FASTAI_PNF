```python
# !pip install -Uqq fastbook
import fastbook
fastbook.setup_book()

from fastbook import *
# !pip install -q jmd_imagescraper
import os
import glob
# from jmd_imagescraper.core import *

import shutil
```

```python
path = Path()
path.ls(file_exts='.pkl')
#Making inferences using our trained model
inferences = load_learner(path/'export.pkl')
```

```python

#Converting the ML model to Web Interface

upload_img_button = widgets.FileUpload()
btn_run = widgets.Button(description='Classify')
out_pl = widgets.Output()

lbl_pred = widgets.Label()

def on_click_classify(change):
  img = PILImage.create(upload_img_button.data[-1])
  out_pl.clear_output()
  with out_pl: display(img.to_thumb(224,224))
  pred,pred_idx,probs = inferences.predict(img)
  lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

btn_run.on_click(on_click_classify)
```

```python
from ipywidgets import *
VBox([widgets.Label('Select your fruit!'),upload_img_button, btn_run, out_pl, lbl_pred])
```
